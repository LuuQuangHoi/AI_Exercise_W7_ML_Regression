```python
import pandas as pd
import matplotlib.pyplot as plt

# Doc file CSV
url = '/content/passenger.csv'
data = pd.read_csv(url)

# Hien thi bieu do Histogram
data.hist()
plt.show()
```

```python
import seaborn as sns
# Tuong quan
correlation = data.corr()
print(correlation)

# Bang phan phoi va tuong quan dung thu vien seaborn
sns.heatmap(correlation,annot = True)
sns.pairplot(data)

```

```python
from sklearn import linear_model 
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Miles(X)']),y = data['Dollars (Y)'])

print(model.intercept_)  # Intercept cua mo hinh
print(model.coef_) # Coffient cua mo hinh
print(model.score(X=pd.DataFrame(data['Miles(X)']),y = data['Dollars (Y)'])) # Do chinh xac cua mo hinh
# y = coef*x + intercept
```

```python
import numpy as np

# Mo hinh hoi quy tuyen tinh
model = linear_model.LinearRegression()
X=pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2,data['Miles(X)']**3,data['Miles(X)']**4]).T
model.fit(X,y = data['Dollars (Y)'])

print('Intercept cua mo hinh:',model.intercept_)
print('Coffient cua mo hinh:',model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X,y= data['Dollars (Y)']))
```

```python
range = np.arange(246,5439,0.1)
data.plot(kind = 'scatter',x='Miles(X)',y='Dollars (Y)',figsize=(9,9),color ='black')
predictor = pd.DataFrame([range ,range**2,range**3,range**4]).T
y_pred = model.predict(X = predictor)
plt.plot(range,y_pred,color = 'blue') 
```
