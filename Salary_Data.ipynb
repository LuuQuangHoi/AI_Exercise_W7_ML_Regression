```python
import pandas as pd
import matplotlib.pyplot as plt

# Doc file CSV
url = '/content/Salary_Data.csv'
data = pd.read_csv(url)

# Hien thi bieu do Histogram
data.hist()
plt.show()

# Do thi phan bo
data.plot(kind = 'scatter',x='YearsExperience',y='Salary',figsize=(9,9),color ='black')
```

```python
import seaborn as sns
correlation = data.corr()

# Bang phan phoi va tuong quan dung thu vien seaborn
sns.heatmap(correlation,annot = True)
sns.pairplot(data)
```

```python
plt.plot(data['YearsExperience'],data['Salary'],'r.')
plt.xlabel('x')
plt.ylabel('y')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model 

# Mo hinh hoi quy tuyen tinh
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['YearsExperience']),y = data['Salary'])

print(model.intercept_)    # Intercept cua mo hinh
print(model.coef_)   # Coffient cua mo hinh
print(model.score(X=pd.DataFrame(data['YearsExperience']),y = data['Salary']))   # Do chinh xac cua mo hinh
```

```python
# Hoi quy Bac 1

# Ve phan bo
data.plot(kind = 'scatter',x='YearsExperience',y='Salary',figsize=(9,9),color ='black')
y_pred = model.predict(X=pd.DataFrame(data['YearsExperience']))

# Ve duong tuyen tinh
plt.plot(data['YearsExperience'],y_pred,color = 'red')
```

```python
# Hoi quy Bac 7
model = linear_model.LinearRegression()
X=pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2,data['YearsExperience']**3,
                data['YearsExperience']**4,data['YearsExperience']**5,data['YearsExperience']**6,
                data['YearsExperience']**7]).T
model.fit(X,y = data['Salary'])
print('Intercept cua mo hinh:',model.intercept_)
print('Coffient cua mo hinh:',model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X,y= data['Salary']))
```

```python
import numpy as np
data.plot(kind = 'scatter',x='YearsExperience',y='Salary',figsize=(9,9),color ='black')
range = np.arange(1.1,10.5,0.1)
predictor = pd.DataFrame([range ,range**2,range**3,range**4,range**5,range**6,range**7]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred,color = 'blue') 
```
